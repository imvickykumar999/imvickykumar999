```python
import pandas as pd

dataframe = pd.read_excel('final.xlsx')
dataframe.index.name = 'S.No'
dataframe.index += 2

columns = ['FULL NAME ( Team Leader )',
           'Contact Number ( Team Leader )',
           'College Name ( All team members must belong from same College )',
           'Team Name',
           'EMAIL ID ( Team Leader )',
           'Full Name ( Member 1 )',
           'Contact Number ( Member 1 )',
           'Full Name ( Member 2 )',
           'Contact Number ( Member 2 )',
           'Full Name ( Member 3 )',
           'Contact Number ( Member 3 )',
          ]   

selrow = [10, 13, 15, 17, 20, 23, 24, 25, 27, 28, 31, 34, 38, 39, 41, 42, 44, 46, 47, 49]
selteam = dataframe.loc[selrow, columns] 

total = set([i for i in range(2, 51)])
unsel = list(total.difference(set(selrow)))
unselteam = dataframe.loc[unsel, columns] 

import pandas as pd
writer = pd.ExcelWriter('Shortlisted Teams.xlsx', engine='xlsxwriter')

selteam.to_excel(writer, sheet_name='Selected')
unselteam.to_excel(writer, sheet_name='Not Selected')

writer.save()
selteam
```
